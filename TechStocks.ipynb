```python
# Finding Min and Max
# print(AAPL_df.loc[AAPL_df['Low'].idxmin()])
# print(AAPL_df.loc[AAPL_df['Close'].idxmax()])

#Locate intial price and final price, use df.shape to locate final close price
apple_initial      = AAPL_df.loc[0][3]
apple_final        = AAPL_df.loc[457][3]

amazon_initial     = AMZN_df.loc[0][3]
amazon_final       = AMZN_df.loc[308][3]

alibaba_initial    = BABA_df.loc[0][3]
alibaba_final      = BABA_df.loc[100][3]

salesforce_initial = CRM_df.loc[0][3]
salesforce_final   = CRM_df.loc[2998][3]

facebook_inital    = FB_df.loc[0][3]
facebook_final     = FB_df.loc[2398][3]

google_initial     = GOOG_df.loc[0][3]
google_final       = GOOG_df.loc[2998][3]



def percent_increase(initial, final):
    """Gives percent incresed from initial price of stock to final price"""
    percentage = 100 * ((final - initial)/ initial)
    print(str(round(percentage,3))+"%" + " Increased from " + str(initial)+ " To "+
     str(final))


percent_increase(apple_initial,apple_final)
print("APPLE")
percent_increase(amazon_initial,amazon_final)
print("AMAZON")
percent_increase(alibaba_initial,alibaba_final)
print("ALIBABA")
percent_increase(salesforce_initial,salesforce_final)
print("SALESFORCE")
percent_increase(facebook_inital,facebook_final)
print("FACEBOOK")
percent_increase(google_initial,google_final)
print("GOOGLE")
```
